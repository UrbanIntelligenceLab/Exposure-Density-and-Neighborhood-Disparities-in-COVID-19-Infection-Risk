__Input variables (zipcode level):__

* Non-residential activity volume change

* Residential activity proportion change 
* Non-residential activity proportion change

# 1. Data preparation
### (1) Load data

```python
data = pd.read_csv('../output/activity_change_before_after_covid_zip.csv', dtype={'MODZCTA':str})
data = data[data['MODZCTA']!='99999'] # drop invalid zipcode
data = data.fillna(0)
data.head()
```

### (2) Create X (input) variables as numpy array

```python
df = data[['MODZCTA', 'delta_R_norm', 'delta_R_pct_norm', 'delta_NR_norm', 'delta_NR_pct_norm', 'delta_O_norm', 'delta_O_pct_norm', 
          'avg_R_pre','avg_O_pre', 'avg_NR_pre', 'avg_R_post','avg_O_post', 'avg_NR_post']]
df.head()
```

```python
# Convert dataframe to numpy array
myarray = df.values[:,1:7]
print (myarray.shape)
# print (myarray)
```

```python
# Data is already set as percentile scale, so not require a normalization process
myarray_norm = myarray
```

# 2. Agglomerative (hierarchical) clustering
### (1) Check dendrogram to select K

```python
# Ward linkage
plt.figure(figsize=(14,4))
d = sch.dendrogram(sch.linkage(myarray_norm, method='ward'))
plt.ylabel('Distance', fontsize=15)
plt.title('Dendrogram - Ward linkage', fontsize=15)
plt.xticks([],[])
# plt.savefig('../figures/dendrogram_ward.pdf', dpi=600)
plt.show()

# Complete linkage
plt.figure(figsize=(14,4))
d = sch.dendrogram(sch.linkage(myarray_norm, method='complete'))
plt.ylabel('Distance', fontsize=15)
plt.title('Dendrogram - Complete linkage', fontsize=15)
plt.xticks([],[])
# plt.savefig('../figures/dendrogram_complete.pdf', dpi=600)
plt.show()

# Average linkage
plt.figure(figsize=(14,4))
d = sch.dendrogram(sch.linkage(myarray_norm, method='average'))
plt.ylabel('Distance', fontsize=15)
plt.title('Dendrogram - Average linkage', fontsize=15)
# plt.savefig('../figures/dendrogram_average.pdf', dpi=600)
plt.xticks([],[])
plt.show()
```

### (2) Clustering using the Ward linkage and K=5 

```python
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

model.fit(myarray_norm)
labels = model.labels_

df['label'] = labels
```

# 4. Contexualize neighborhood chatacteristics and indentify disparities in behavioral responses to social distancing

### 1) Data preparation

```python
# Load explanatory variables
ch = pd.read_csv('../output/independenta_variables.csv', dtype={'MODZCTA':str})
# Merge with the clustering result
re = pd.merge(df[['MODZCTA', 'label']], ch, on='MODZCTA', how='left')
re = re[re['MODZCTA']!='99999']
re = re.fillna(0)
re.head(2)
```

```python
# Add Hispanic information that was not included in the previous data collection stage
his = pd.read_csv('../data/acs_2018_hispanic.csv', skiprows=[1], dtype={'ZIP Code Tabulation Area (5-digit)':str})
his = his[['ZIP Code Tabulation Area (5-digit)', '% Total Population: Not Hispanic or Latino: White Alone']]
his.columns = ['MODZCTA', 'non_white_hispanic']
his.head()

re = pd.merge(re, his, on='MODZCTA', how='left')
re.head()
```

### 2) One-way ANOVA

```python
anova_value = []
anova_pvalue = []
sig_col = []
for col in re.columns.tolist()[2:]:
#     print (col)
    inc2 = re[col][re['label']==4].values
    inc1 = re[col][re['label']==2].values
    stable = re[col][re['label']==1].values
    dec1 = re[col][re['label']==3].values
    dec2 = re[col][re['label']==0].values
#     print (stats.f_oneway(inc2, inc1, stable, dec1, dec2)[0].round(4))
    anova_value.append(stats.f_oneway(inc2, inc1, stable, dec1, dec2)[0].round(4))
#     print (stats.f_oneway(inc2, inc1, stable, dec1, dec2)[1].round(4))
    anova_pvalue.append(stats.f_oneway(inc2, inc1, stable, dec1, dec2)[1].round(4))
#     print ('-------------------------')
    if stats.f_oneway(inc2, inc1, stable, dec1, dec2)[1].round(4) <= 0.05:
        sig_col.append(col)
```

```python
# Re-create label names with colors
re['label_'] = 'Orange'
re['label_'][re['label']==0] = 'Green'
re['label_'][re['label']==1] = 'Yellow'
re['label_'][re['label']==2] = 'Red'
re['label_'][re['label']==3] = 'Blue'
```

### 3) Tukey test

```python
# Tukey test
for col in sig_col:
#     print ("------------------", col, "--------------------")
    mcDate = multi.MultiComparison(re[col], re['label_'])
    Results = mcDate.tukeyhsd()
#     print (Results)
```

### 4. Create a result table to contexualize neighborhood characteristics of each cluster group

```python
mean_std_list_total = []
for col in sig_col:
    mean_std_list = []
    for i in [0,1,2,3,4]:
        group = ['Green', 'Yellow', 'Red', 'Blue', 'Orange']
        mean = re[col][re['label']==i].mean()
        std = re[col][re['label']==i].std()
        mean_std = round(mean,4), round(std,4)
        mean_std_list.append(mean_std)
        
    mean_std_list_total.append(mean_std_list)
```

```python
# print (len(mean_std_list_total))
# print (len(mean_std_list_total[0]))
# print (mean_std_list_total[0])

df_summary = pd.DataFrame(mean_std_list_total)
df_summary['feature'] = sig_col
df_summary.columns = ['Green', 'Yellow', 'Red', 'Blue', 'Orange', 'feature']
df_summary = df_summary[['feature', 'Green', 'Yellow', 'Red', 'Blue', 'Orange']]

for c in ['Green', 'Yellow', 'Red', 'Blue', 'Orange']:
    df_summary[c] = df_summary[c].astype(str).str.replace('(','')
    df_summary[c] = df_summary[c].astype(str).str.replace(', ',' (')
    
df_summary.head()
```
